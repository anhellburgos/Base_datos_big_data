# OPERACIONES DATASET PARA NEO4J

```python
import pandas as pd
#SE IMPORTA DE GITHUB CSV NORMALIZADO
path = "https://raw.githubusercontent.com/anhellburgos/Base_datos_big_data/main/act-grupal-openDataLocalesMadrid-corregido.csv"
df = pd.read_csv(path)

print(df.shape) 
```

```python
#RENOMBRAMOS LA COLUMNA _ID PARA QUE SEA MÁS ACORDE A SU CONTENIDO
df.rename(columns={'_id': 'id_terraza'}, inplace=True)
```

```python
#REDUCIMOS LA CANTIDAD DE REGISTROS DEBIDO A LOS RECURSOS DE HARDWARE Y EL USO DE NEO4J
df = df.loc[1:100]
```

```python
#GENERAMOS CADA UNA DE LOS SET DE DATOS PARA CADA ENTIDAD SEGUN ANÁLISIS PREVIO
#ENTIDAD TERRAZA
df_Terraza =  df[["id_terraza","Superficie_ES","hora_ini_LJ_es","hora_fin_LJ_es",
                  "hora_ini_LJ_ra","hora_fin_LJ_ra","hora_ini_VS_es","hora_fin_VS_es",
                  "hora_ini_VS_ra","hora_fin_VS_ra","mesas_aux_es","mesas_aux_ra",
                  "mesas_es","mesas_ra","sillas_es","sillas_ra","DESC_CLASE",
                  "DESC_NOMBRE","num_terraza","cal_terraza","desc_ubicacion_terraza"]]
df_Terraza.to_csv(r'D:/NEO4J/Terrazas.csv', index=False)
```

```python
#ENTIDAD LOCAL
df_Local =  df[["id_local","desc_situacion_local","Nombre","clase_vial_edificio",
                "num_edificio","Cod_Postal","coordenada_x_local","coordenada_y_local"]]

df_Local.to_csv(r'D:/NEO4J/Local.csv', index=False)
```

```python
#ENTIDAD PERIODO ESTACIONAL
df_Periodo_Estacional =  df[["id_periodo_terraza", "desc_periodo_terraza"]]
df_Periodo_Estacional.to_csv(r'D:/NEO4J/Periodo_Estacional.csv', index=False)
```

```python
#ENTIDAD SITUACIÓN TERRAZA
df_Situacion_Terraza =  df[["id_situacion_terraza","desc_situacion_terraza"]]
df_Situacion_Terraza.to_csv(r'D:/NEO4J/Situacion_Terraza.csv', index=False)
```

```python
#ENTIDAD MUNICIPIO
df_Municipio =  df[["id_distrito_local", "desc_distrito_local"]]
df_Municipio.to_csv(r'D:/NEO4J/Municipio.csv', index=False)
```

```python
#ENTIDAD BARRIOS
df_Barrios =  df[["desc_barrio_local"]]
df_Barrios.to_csv(r'D:/NEO4J/Barrios.csv', index=False)
```

```python
#GENERAMOS CADA UNA DE LAS RELACIONES EXISTENTES
#TERRAZA - LOCAL
df_Contiene =  df[["id_terraza", "id_local"]]
df_Contiene.to_csv(r'D:/NEO4J/R_Contiene.csv', index=False)
```

```python
#LOCAL - MUNICIPIO
df_Pertenece =  df[["id_local", "id_distrito_local"]]
df_Pertenece.to_csv(r'D:/NEO4J/R_Pertenece.csv', index=False)
```

```python
#MUNICIPIO - BARRIOS
import numpy as np

df_EsContenedor= df[["id_distrito_local", "desc_barrio_local"]] 
df_EsContenedor = df_EsContenedor.drop_duplicates(['id_distrito_local','desc_barrio_local'])[['id_distrito_local','desc_barrio_local']]
df_EsContenedor.to_csv(r'D:/NEO4J/R_EsContenedor.csv', index=False)
```

```python
#TERRAZA - SITUACION TERRAZA
df_TieneSituacion =  df[["id_terraza","id_situacion_terraza"]]
df_TieneSituacion.to_csv(r'D:/NEO4J/R_TieneSituacion.csv', index=False)
```

```python
#TERRAZA - PERIODO ESTACIONAL
df_Esta =  df[["id_terraza","id_periodo_terraza"]]
df_Esta.to_csv(r'D:/NEO4J/R_Esta.csv', index=False)
```
